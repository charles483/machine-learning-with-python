##  the  main aim of this exercise is to predict the brain size given the  head size

#### load the required modules 

```python
import numpy as np
import seaborn
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd 
seaborn.set()
```

### loading the data

```python
data =pd.read_csv('headbrain11.csv')
```

```python
data
```

```python
data.describe()
```

###  exploring the data

```python
y=data['Brain Weight(grams)']
x1=data['Head Size(cm^3) ']
```

```python
plt.scatter(x1,y)
plt.xlabel('Head Size(cm^3) ', fontsize=15)
plt.ylabel('Brain Weight(grams)', fontsize=15)
plt.show()
```

### The regression itself

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat=0.2634	*x1 +325.5734	
fig=plt.plot(x1, yhat, c='blue', lw=4, label='regression line')
plt.xlabel('Head Size(cm^3) ', fontsize=15)
plt.ylabel('Brain Weight(grams)', fontsize=15)
plt.show()
```

### let us try to predict the brain size

```python
new_data = pd.DataFrame({'const':1, 'Head Size(cm^3)': [4512]}) 
predicted_brain_size=results.predict(new_data)
```
