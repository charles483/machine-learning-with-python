# Here we are predicting profits using stats model

#### importing the necessary modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn
import statsmodels.api as sm
seaborn.set()
```

## loading the  and explorint the data

```python
# loading the data 
data = pd.read_csv('50_Startups.csv')
```

```python
data
```

```python
data.describe()
```

```python
y=data['Profit']
x1=data['Marketing Spend ']
```

```python
plt.scatter(x1, y)
plt.xlabel('Profits', fontsize=14)
plt.ylabel('Marketing Spend', fontsize=14)
plt.title('Profits vs Marketing Spend')
plt.show()
```

### regression 

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1, y)
yhat=0.2465	*x1 +6e+04	
fig=plt.plot(x1, yhat, c='red',lw=2, label='regression line' )
plt.xlabel('Profits', fontsize=14)
plt.ylabel('Marketing Spend', fontsize=14)
plt.title('Profits vs Marketing Spend')
plt.show()
```
