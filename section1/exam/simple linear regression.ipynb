```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data =pd.read_csv('startups.csv')
```

### defining the dependent variable

```python
y = data['Profit']
x1 = data[['Marketing Expenditure','R&D Expenditure']]
```

```python
# # exploring the data through graphs
# plt.scatter(x1, y)
# plt.xlabel('Profit',fontsize=20)
# plt.ylabel('Marketing Expenditure',fontsize=20)
# plt.show()
```

```python
# crreating the regression itself using the stats model

x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(x1,y)
yhat=0.0017*x1 +0.275
fig = plt.plot(x1, yhat, lw=4 , c='orange',label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```
